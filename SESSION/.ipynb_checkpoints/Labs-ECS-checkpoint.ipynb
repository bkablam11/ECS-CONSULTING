#### Lab 3 : Graphe de fonctions

```python
from turtle import *
from math import *

# square function
speed("fastest")
width(2)
color('blue')
up()

for x in range(-200,200):
    if x == -199: 
        down()
    y = 1/ 100 * x ** 2   # Parabole
    goto(x,y)

# sinus function
speed("fastest")
width(2)
color('red')
up()

for x in range(-200,200):
    if x == -199: 
        down()
    y = 100*sin(1/20*x)     # Sinus
    goto(x,y)

exitonclick()
```

```python
# Préparation

from turtle import *

tortue1 = Turtle()
tortue2 = Turtle()
tortue3 = Turtle()
tortue4 = Turtle()

tortue1.speed("fastest")
tortue2.speed("fastest")
tortue3.speed("fastest")
tortue4.speed("fastest")

tortue1.color('red')
tortue2.color('blue')
tortue3.color('orange')
tortue4.color('green')


tortue1.up()
tortue1.goto(-200,-200)
tortue1.down()

tortue2.up()
tortue2.goto(200,-200)
tortue2.down()

tortue3.up()
tortue3.goto(200,200)
tortue3.down()

tortue4.up()
tortue4.goto(-200,200)
tortue4.down()


# Boucle principale

for i in range(40):

    position1 = tortue1.pos()
    position2 = tortue2.pos()
    position3 = tortue3.pos()
    position4 = tortue4.pos()

    tortue1.goto(position2) 
    tortue1.goto(position1)       

    tortue2.goto(position3)
    tortue2.goto(position2)

    tortue3.goto(position4)
    tortue3.goto(position3)

    tortue4.goto(position1)
    tortue4.goto(position4)

    angle1 = tortue1.towards(position2)  
    tortue1.setheading(angle1)   

    angle2 = tortue2.towards(position3)
    tortue2.setheading(angle2)

    angle3 = tortue3.towards(position4)
    tortue3.setheading(angle3)

    angle4 = tortue4.towards(position1)
    tortue4.setheading(angle4)

    tortue1.forward(10)    
    tortue2.forward(10)
    tortue3.forward(10)
    tortue4.forward(10)

exitonclick()
```

### TP 2 - Lab 5 (Statistique de base)

```python
def sum2(liste):
    sum22 = 0
    for i in range(len(liste)):
        sum22+=liste[i]
    return sum22

#test
# 1
liste1=[4,8,3]
print('result of my function sum',sum2(liste1))
print('result of function sum python',sum(liste1))
```

```python
# 2
# https://www.tutorialspoint.com/custom-len-function-in-python
## function to calculate lenght of the iterator
def length(iterator):
   ## initializing the count to 0
   count = 0
   ## iterating through the iterator
   for item in iterator:
      ## incrementing count
      count += 1
   ## returning the length of the iterator
   return count

# test 
liste1=[4,8,3]
#length(liste1)
def moyenne(liste):
    if length(liste)==0:
        return 0
    else:
        return sum2(liste)/length(liste)

moyenne(liste1)
```

```python
# 3
def minimum(liste):
    for i in range(1, len(liste)):
        mini=liste[0]
        if liste[i]<mini:
            mini=liste[i]
    return mini

liste1=[10, 5, 9, 12]
print("result of my min function", minimum(liste1))
print("result of python min function", min(liste1))
print("result of python min function", min(12,7))
```

```python
## Question 4 ##
def variance(liste):
    if length(liste)==0:
        var = 0
    else:
        sum_square = 0
        for x in liste:
            sum_square = (sum_square + (x-moyenne(liste))**2)
            var = sum_square/length(liste)
    return var

liste1 = [6, 8, 2, 10]
print(f"Moyenne de la  serie {liste1}: {moyenne(liste1)}")
print(f"Variance de la serie {liste1} : {variance(liste1)}")

## Question 5 ##
def ecart_type(liste):
    from math import sqrt
    e = sqrt(variance(liste))
    return e

print(f"l'ecart type de la serie {liste1} : {ecart_type(liste1)}")
```
