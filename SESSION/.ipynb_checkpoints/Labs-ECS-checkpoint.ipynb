#### Lab 1 : Drawing P Y

```python
# import library

from turtle import *
import turtle

width(5)    # Epaisseur du trait

# Lettre "P"

color('red')

left(90)        # 90 degrés à gauche
forward(200)    # On avance

right(90)
forward(100)

right(90)
forward(100)

right(90)
forward(100)

up()

# Fin du code 

# Lettre "Y"

color('blue')

goto(200,0)
down()
setheading(90)
forward(120)
left(30)
forward(75)
backward(75)
right(60)
forward(75)

exitonclick()
```

#### Lab 2 : Figures Géométriques

```python
from turtle import *

# Un pentagone
up()
goto(-450, 0)
down()

width(5)
color('blue')

for i in range(5):
    forward(100)
    left(72)


##############################

# Un autre pentagone
up()
goto(-250, 0)
down()
color('red')

longueur = 200
angle = 72
for i in range(5):
    forward(longueur)
    left(angle)


##############################
# Question 3

# Un dodecagone (12 côtés quoi)
up()
goto(250, 0)
down()

color("purple")
n = 12
angle = 360/n
for i in range(n):
    forward(100)
    left(angle)


##############################
# Question 4

# Une spirale
up()
goto(-450, -150)
down()

color("green")
longueur = 2
for i in range(25):
    forward(longueur)
    left(40)
    longueur = longueur + 2

exitonclick()
```

#### Lab 3 : Graphe de fonctions

```python
from turtle import *
from math import *

# square function
speed("fastest")
width(2)
color('blue')
up()

for x in range(-200,200):
    if x == -199: 
        down()
    y = 1/ 100 * x ** 2   # Parabole
    goto(x,y)

# sinus function
speed("fastest")
width(2)
color('red')
up()

for x in range(-200,200):
    if x == -199: 
        down()
    y = 100*sin(1/20*x)     # Sinus
    goto(x,y)

exitonclick()
```

#### Lab 4 : Triangle de Sierpinski

```python
from turtle import *

width(5)
up()
goto(-400,100)
down()

# level 1
for i in range(3):
    color("blue")
    forward(256)
    left(120)

width(5)
up()
goto(-100,-100)
down()
# level 2
for i in range(3):
    color("blue")
    forward(256)
    left(120)

    for i in range(3):
        color("red")
        forward(128)
        left(120)

width(5)
up()
goto(300,-100)
down()        
# level 3 
for i in range(3):
    color("blue")
    forward(256)
    left(120)

    for i in range(3):
        color("red")
        forward(128)
        left(120)

        for i in range(3):
            color("green")
            forward(64)
            left(120)
            
# level 4
width(5)
up()
goto(-400,-200)
down()  
for i in range(3):
    color("blue")
    forward(256)
    left(120)

    for i in range(3):
        color("red")
        forward(128)
        left(120)

        for i in range(3):
            color("green")
            forward(64)
            left(120)  

            for i in range(3):
                color("orange")
                forward(32)
                left(120)     

exitonclick()
```
